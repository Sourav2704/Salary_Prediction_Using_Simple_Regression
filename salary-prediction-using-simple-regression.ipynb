# Import Libraries

```python
# Importing the important Libraries
import seaborn as sns
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

# Import Dataset

```python
# Importing the Datasets
df=pd.read_csv("/kaggle/input/salary-prediction-data-simple-linear-regression/Salary_Data.csv")
df.info()
```

# Assumptions
- 1. Linear Relationship between dependent and independent variable
- 2. Both variables are numeric
- 3. Both variables have homosceaticty
- 4. Independent variables are not correlated with each other

```python
model=LinearRegression()


X=df[["YearsExperience"]]

y=df["Salary"]

model.fit(X ,y)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
X_train,X_test,y_train,y_test= train_test_split(X,y,random_state=1)
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
plt.plot(X_test,y_pred,color='red')
plt.scatter(X,y,color='green')
plt.scatter(X_test,y_test,color='blue')
```

# R2_Score & MAE MSE

```python
print("R2_score: ", r2_score(y_test,y_pred))
print("MAE: ", mean_absolute_error(y_test,y_pred))
print("MSE: ", mean_absolute_error(y_test,y_pred))
```

# Interpretation of result

### Simple Linear Regression is a statistical method used to model the relationship between two variables whose
###  one variable is the predictor or independent and other variable is response or dependent variable


   - The R2 score is known as coefficient of determination .it measures how good the model is fit
   - It ranges From 0 to 1
   - where 1 value indicates a perfect fit
   - o value indicates no fit at all
   - In this case R2_score value is  0.911147941293075512.which is perfect fit
   - This means that the model is  able to capture significant amount of variation of data
   - it will be goot fit in this model at all
   
   
   -  The MAE is a measure of the average absolute difference between the predicted and actual values of response variable
   -  MSE measures the average squared difference between the predicted values and the actual values.
   -  In this case, the low values of MAE and MSE indicate that, on average, the predicted values are very close to the actual values.
